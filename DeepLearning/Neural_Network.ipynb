```python
import numpy as np
import matplotlib.pylab as plt
```

# 활성화 함수
* 퍼셉트론에서 입력 신호의 총합을 출력으로 바꾸어주는 함수를 **활성화 함수**라고 한다.
* 단층(단순) 퍼셉트론은 1 또는 0의 출력만을 가질 수 있다. 즉 단순 퍼셉트론의 활성화 함수는 1 또는 0의 출력만을 가진다. 
* 이러한 활성화 함수를 **계단 함수**라고 한다. 이는 활성화 함수의 그래프를 그려보았을 때 계단과 같은 모양이 그려지는 것에서 유래한다.
* 활성화 함수를 계단 함수에서 다른 함수로 변경하는 것(==퍼셉트론을 다층으로 쌓는 것)이 신경망의 세계로 나아가는 열쇠이다!
* * *

## 계단 함수

```python
def step_function(x):
    return np.array(x>0, dtype=int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) #y축의 범위 지정
plt.show()
```

## 시그모이드 함수

**h(x) = 1/(1+exp(-x))**

* 신경망에서 자주 사용하는 **시그모이드 함수**이다.
* 신경망에서는 활성화 함수로 시그모이드 함수를 사용하여 신호를 변환하고 그 변환된 신호를 다음 뉴런으로 전달한다.
* **퍼셉트론과 신경망의 차이는 이 활성화 함수 뿐이다.**
* 시그모이드 함수와 계단 함수의 가장 눈에 띄는 차이는 매끄러움이다. 이 **매끈함**이 신경망 학습에서 아주 중요한 역할을 할 것이다.

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.array([-1.0, 1.0, 2.0])
sigmoid(x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
y1 = step_function(x)
y2 = sigmoid(x)
plt.plot(x, y1, linestyle="--")
plt.plot(x, y2)
plt.show()
```

## 비선형 함수
* 계단 함수와 시그모이드 함수는 모두 **비선형 함수**이다.
* 선형 함수는 출력이 입력의 상수배 만큼 변하는 함수를 말한다. 수학의 일차함수와 같다.
* 비선형 함수는 그래프를 직선 하나로는 그릴 수 없는 함수이다.
***

## ReLU 함수
**h(x) = x if x>0 else 0**
* 0 초과의 입력에는 입력을 그대로 출력하고, 0 이하의 입력에는 0을 출력하는 함수(비선형 함수)이다.
* 신경망 분야에서는 시그모이드 함수가 오랬동안 이용되어 왔으나 최근에는 렐루 함수를 주로 이용한다.

```python
def relu(x):
    return np.maximum(x, 0)

y = relu(x)
plt.plot(x, y)
plt.show()
```
